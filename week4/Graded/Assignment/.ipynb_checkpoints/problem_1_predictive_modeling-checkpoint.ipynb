### We are gonna replace the null value of depth with median()

```python
df.drop(df[(df['x'] == 0) | (df['y'] == 0) | (df['z'] == 0)].index , inplace = True)
```

```python
df[(df['x'] == 0) | (df['y'] == 0) | (df['z'] == 0)]
```

```python
df['depth'] = df['depth'].fillna(df['depth'].median())
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import StandardScaler

X = StandardScaler()

num_df = df.select_dtypes(include = ['int64','float64'])

df[num_df.columns] = X.fit_transform(num_df)
```

```python
df.head()
```

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
for column in df[num_df.columns].columns:
    lr,ur=remove_outlier(df[column])
    df[column]=np.where(df[column]>ur,ur,df[column])
    df[column]=np.where(df[column]<lr,lr,df[column])
```

### SPLIT

```python
# Copy all the predictor variables into X dataframe
X = data.drop('price', axis=1)

# Copy target into the y dataframe. 
y = data[['price']]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3 , random_state=1)
```

### Linear Regression

```python
regression_model = LinearRegression()
regression_model.fit(X_train,y_train)
```

```python
for idx, col_name in enumerate(X_train.columns):
    print("The coefficient for {} is {}".format(col_name, regression_model.coef_[0][idx]))
```

```python
intercept = regression_model.intercept_[0]

print("The intercept for our model is {}".format(intercept))
```

```python
regression_model.score(X_train, y_train)
```

```python
regression_model.score(X_test, y_test)
```

```python
predicted_train=regression_model.fit(X_train, y_train).predict(X_train)
np.sqrt(metrics.mean_squared_error(y_train,predicted_train))
```

```python
predicted_test=regression_model.fit(X_train, y_train).predict(X_test)
np.sqrt(metrics.mean_squared_error(y_test,predicted_test))
```

```python
intercept = regression_model.intercept_[0]

print("The intercept for our model is {}".format(intercept))
```

```python
# R square on training data
regression_model.score(X_train, y_train)
```

```python
#RMSE on Testing data
predicted_test=regression_model.fit(X_train, y_train).predict(X_test)
np.sqrt(metrics.mean_squared_error(y_test,predicted_test))
```

## Linear Regression using statsmodels

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

data_train = pd.concat([X_train, y_train], axis=1)
data_test = pd.concat([X_test, y_test] , axis = 1)
data_train.head()
```

```python
data_train.columns
```

```python
data_train.rename(columns = {'cut_Very Good' : 'cut_Very_Good'}, inplace = True)
data_test.rename(columns = {'cut_Very Good' : 'cut_Very_Good'}, inplace = True)
```

```python
data_train.columns
```

```python
import statsmodels.formula.api as smf

expr = 'price ~ carat + depth + table + x+ y + z + cut_Good + cut_Ideal + cut_Premium + cut_Very_Good + color_E + color_F + color_G + color_H + color_I + color_J + clarity_IF + clarity_SI1 + clarity_SI2 + clarity_VS1 + clarity_VS2 + clarity_VVS1 + clarity_VVS2'

lm1 = smf.ols(formula= expr, data = data_train).fit()
lm1.params
```

```python
print(lm1.summary())
```

```python
mse = np.mean((lm1.predict(data_train.drop('price',axis=1))-data_train['price'])**2)
```

```python
np.sqrt(mse)
```

```python
np.sqrt(lm1.mse_resid)
```

```python
y_pred = lm1.predict(data_test)
```

```python
plt.scatter(y_test['price'], y_pred)
plt.show()
```
